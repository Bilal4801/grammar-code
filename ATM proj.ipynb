# ATM Project

```python
# make an atm project
# first import time
import time
print("please insert your amount:")

time.sleep(2)
password = 8521
pin = int(input('enter the pin:'))
balance = 30000
# balance.max() = 30000
# balance.min() = 1000


if pin == password:
    while True:
        print("""
        1 == check balance
        2 == deposit balance
        3 == withdraw balance
        4 == pay utility bills
        5 == exit
        """)
        try:
            option = int(input("please enter your choice:"))
        except:
            print('please enter valid option:')

        if option == 1:
            print(f"(your current balance is {balance}.)")
            print("==============================================")
            print("==============================================")
            print("==============================================")
            break
        if option == 2:
            deposite_amount = int(input("enter the deposite amount:"))
            print("==============================================")
            print("==============================================")
            balance += deposite_amount
            
            print("==============================================")
            print("==============================================")
            print(f"({deposite_amount} is credited from your account:)")
            print("==============================================")
            print("==============================================")
            
            print(f"(your update balance is {balance})")
            break
        if option == 3:
            withdraw_amount = int(input('enter the amount:'))
            balance -= withdraw_amount
            print("==============================================")
            print("==============================================")
            
            print(f"({withdraw_amount} is debited from your acount)")
            
            print("==============================================")
            print("==============================================")
            
            print(f"(your current balance is {balance})")
            print("==============================================")
            print("==============================================")
            break

#             print(f'(your update balance is {balance})')
#             print("==============================================")
#             print("==============================================")

        if option == 4:
            pay_utility = int(input("enter the amount:"))
            balance -= pay_utility
            print("==============================================")
            print("==============================================")
            print(f'({pay_utility}you pay utility bill.\n Thank you')
            print("==============================================")
            print("==============================================")

        if option == 5:
            break

else:
    print("wrong pin please try again")
```
